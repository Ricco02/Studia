# Matplotlib.pyplot - podstawy

Biblioteka Matplotlib służy do tworzenia różnych typów wykresów i wizualizacji danych w 2D. Do tworzenia różnych typów wykresów w stylu MatLaba służy biblioteka Pyplot, będąca częścią Matplotlib. ## Plot

Podstawową funkcją w Pyplot jest `plot()`, gdzie argumentami są typowo dwie listy (lub jednowymiarowe macierze `ndarray`, co jest dużo wygodniejsze w praktyce) `x` i `y` zawierające współrzędne $x$ i $y$ punktów wykresu, odpowiednio.

Prosty przykład:

````python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,2,0.01)
y=np.sin(x*np.pi)

plt.plot(x,y)
````

Style wykresu do wyboru:
- `'-'` linia ciągła
- `'--'` linia przerywana
- `'-.'` linia kropka kreska
- `':'` linia wykropkowana
- `'.'` pojedyncze punkty (nie połączone linią)
- `','` pojedyncze piksele
- `'o'` pojedyncze kółka
- `'v'`, `'^'`, `'<'}`, `'>'` trójkąty
- ` '1', '2', '3', '4'` dzióbki
- ` 's'` kwadraty
- ` 'p'` pięciokąty
- ` '*'` gwiazdki
- `'h', 'H'` sześciokąty
- ` '+'` plusy
- ` 'x'` iksy
- ` 'D', 'd'` romby
- ` '|'` linie pionowe
- `'_'` linie poziome

Kolory wykresu do wyboru:
- ` 'b'` niebieski
- ` 'g'` zielony
- ` 'r'` czerwony
- ` 'c'` cyjan
- ` 'm'` magenta
- `'y'` żółty
- `'k'` czarny
- `'w'` biały

Niektóre podstawowe opcje wykresu:
 - ` plt.xlabel('opis')` - podpis pod osią OX
 - ` plt.ylabel('opis')` - podpis pod osią OY
 - ` plt.title('nazwa')` - nazwa całego wykresu
 - ` plt.grid(True/False)` - włącza/wyłącza widoczną siatkę
 - ` plt.axis()` - funkcja do zarządzania osiami, np. `[x1,x2,y1,y2]` zakresy na osiach, `'off'` wyłącza widok osi, `'equal'` równe jednostki na osiach, `'scaled'` zmiejsza ramkę tak, aby jednostki na osiach były równe, `'square'` ustawia równe zakresy na osi OX i OY itd.
 - ` plt.legend(['opis1','opis2'])` - legenda (`'opis2'` itd. dla wielu wykresów)
 - ` plt.arrow(x0,y0,dx,dx)` - rysuje strzałkę o początku w punkcie $(x0,y0)$ i końcu w $(x0+dx,y0+dy)$ (przydatne do szkicowania osi)
 - ` plt.autoscale(enable=True, axis='both', tight=None)` - włącza lub wyłącza autoskalowanie osi. Wybór osi: `'x'`, `'y'`, `'both'`. Opcja `tight` włącza lub wyłącza marginesy naokoło wykresu (przy wyłączonych marginesach wykres będzie ,,dotykał'' ramki).

Przykładowe zastosowanie wybranych opcji:

````python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-2,2.01,0.01)
y=np.sin(x*np.pi)

plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('sin(pi*x)')
plt.title('Przyklad')
plt.grid(True)
plt.axis([-2.5,2.5,-1.1,1.1])
plt.arrow(-2.5,0,5,0,length_includes_head=True,head_width=0.1)
plt.arrow(0,-1.1,0,2.2,length_includes_head=True,head_width=0.1)
````

## Łączenie wykresów

Aby na jednym obrazie uzyskać kilka wykresów po prostu wywołujemy w skrypcie funkcję `plot` żądaną liczbę razy:

````python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-2,2.01,0.01)
y=x**2

plt.plot(x,x)
plt.plot(x,y,'g')
plt.axis('scaled')
plt.grid(True)
plt.legend(['y=x','y=x^2'])
````

Jeżeli chcemy w jednej grafice umieścić kilka wykresów oddzielnie (w formie tabeli wykresów) używamy funkcji `plt.subplot(r,c,n)`, gdzie `r` to liczba wierszy, `c` - liczba kolumn a `n` to liczba porządkowa wykresu. 

````python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-6,6.01,0.01)
t=np.tan(x)
s=np.arctan(x)

plt.subplot(1,2,1)
plt.plot(x,t)
plt.axis([-6,6,-3,3])
plt.grid(True)
plt.title('tg(x)')

plt.subplot(1,2,2)
plt.plot(x,s)
plt.axis([-6,6,-3,3])
plt.grid(True)
plt.title('arctg(x)')
````

## Zadanie 2.

Na jednej grafice w czterech odrębnych wykresach naszkicować cztery podstawowe krzywe stożkowe: okrąg, elipsę, parabolę i hiperbolę (we współrzędnych kartezjańskich). Podpisać każdy wykres nazwą krzywej, wyświetlić siatkę. Każda krzywa powinna być w innym kolorze.

````python
x=np.arange(-6.01,6.01,0.01) 
t= np.linspace(-np.pi,np.pi, 1000) 

#circle
plt.subplot(2,2,1) 
xcircle=np.sin(t) 
ycircle=np.cos(t)
plt.plot(xcircle,ycircle) 
plt.axis("equal")
plt.title("Okrąg") 
plt.grid('True') 

#elipse
plt.subplot(2,2,2) 
xelipse=2*np.sin(t)
yelipse=4*np.cos(t)
plt.plot(xelipse,yelipse,"r")
plt.axis("equal")
plt.title("Elipsa")
plt.grid('True')

#parabola
plt.subplot(2,2,3) 
xparab=x
yparab=x**2
plt.plot(xparab,yparab,"g")
plt.axis("equal")
plt.title("Parabola")
plt.grid('True')

#hyperbola
plt.subplot(2,2,4) 
xhyper=-1*np.cosh(t/2)
yhyper=np.sinh(t/2)
mxhyper=-xhyper
plt.plot(xhyper,yhyper,"y")
plt.plot(mxhyper,yhyper,"y")
plt.axis("equal")
plt.title("Hiperbola")
plt.grid('True')
````

## Zadanie 3.

W umieszczonym na Delcie pliku `kroliki.txt` znajdują się dane dotyczące populacji królików, lisów i marchewek na przestrzeni 20 lat (od 1990 do 2010).
1. Zobrazować dane na wspólnym wykresie liniowym. Dodać legendę, siatkę i adekwatne nazwy osi.
1. Obliczyć i zaznaczyć na wykresie w czytelny sposób maksimum dla danej populacji.
1. Obliczyć średnią populację poszczególnych gatunków.
1. Obliczyć dla każdego roku, który gatunek był najliczniejszy.

````python
data=np.loadtxt("kroliki.txt",skiprows=1) 
plt.plot(data[:,0],data[:,1:]) 
plt.legend(["Króliki","Lisy","Marchewki"]) 
plt.grid("True")
plt.xlabel("Rok")
plt.ylabel("Liczebność")
````

Obliczyć i zaznaczyć na wykresie w czytelny sposób maksimum dla danej populacji.

````python
ind=np.argmax(data[:,1:],axis=0) #obliczenie w którym wierszu znajduje się największa wartość dla danej kolumny
ind
````

````python
xind=[data[ind[i],0] for i in range(len(ind))] # wypisanie lat w których jakakolwiek populacja osiągnęła swoje maksimum
xind
````

````python
yind=[data[ind[i],i+1] for i in range(len(ind))]
plt.plot(data[:,0],data[:,1:]) 
plt.plot(xind,yind,"ro")
plt.legend(["Króliki","Lisy","Marchewki"])
plt.grid("True") 
plt.xlabel("Rok")
plt.ylabel("Liczebność")
for index, value in zip(xind,yind): 
    plt.text(index, value+1300, "%d" %value, ha="center")
````

Obliczyć średnią populację poszczególnych gatunków.

````python
mean=data[:,1:].mean(axis=0).round(4) 
for name,value in zip(["królików","lisów","marchwi"],mean): 
    print(f"Średnia populacja {name} wynosi {value} \n") 
````

Obliczyć dla każdego roku, który gatunek był najliczniejszy.

````python
yrs=np.argmax(data[:,1:],axis=1)
listt=["Króliki" if x==0  else "Lisy" if x==1  else "Marchew" for x in yrs] 
for year,value in zip(data[:,0],listt): 
    print(f"Rok {int(year)} - {value}") 
````
