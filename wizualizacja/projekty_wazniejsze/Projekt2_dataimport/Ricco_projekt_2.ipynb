```python
import pandas as pd
import os
```

```python
#handling bad line

def is_int(x):
   try:
      x=int(x)
   except:
      return False
   return True

def handle_bad_line(line):
   for i in range(len(line)-1):
      if line[i+1] == str(i+1) and not is_int(line[i]):
         line[i] = str(i+1)
         line[i+1] = ''
      
   line = [0 if not is_int(x) else x for x in line[0:9]]
   return line[0:9]
```

```python
def dataMaker(name):
    data_for_index=pd.read_csv(name,nrows=0).columns.values
    index_for_file = pd.date_range(start= data_for_index[0][9:] + ' ' + data_for_index[1][6:], periods=int(data_for_index[2][8:]), freq=data_for_index[3][8:])
    collected_data=pd.read_csv(name,skiprows=1, skip_blank_lines=True, on_bad_lines=handle_bad_line, engine='python')
    collected_data.index=index_for_file
    collected_data.name = str(name[0:-4])+'_df'
    return collected_data
```

```python
def errorHandling(dataframe):
    empty_rows = dataframe[dataframe.isnull().all(axis=1)].index.values
    dataframe = dataframe.dropna(how='all')
    print(f"{len(empty_rows)} empty rows deleted, here are indicies: {empty_rows}")
    mask=dataframe.iloc[:].eq(0).sum(axis=1)>5
    bd_error = dataframe[mask]
    bd_error_handled=dataframe[~mask]
    bd_error_handled=bd_error_handled.replace(0,{col: i+1 for i,col in enumerate(dataframe.columns)})
    return(bd_error,bd_error_handled)
```

```python
def dataCount(dataframe,how):
    for col_ind,col in enumerate(dataframe.columns):
        good_response = dataframe[dataframe[col]==col_ind+1][col].count()  
        bad_response = dataframe[dataframe[col]!=col_ind+1][col].count()
        percent_of_errors=bad_response/(good_response+bad_response)*100
        print(f"{col}\n{percent_of_errors.round(2)}% of errors, no response as error: {how}")
```

```python
file_list=['pomiar_1.csv','pomiar_2.csv','pomiar_3.csv','pomiar_4.csv']
for file_number,file_name in enumerate(file_list):
    dataframe=dataMaker(name=file_name)
    bd_error,bd_error_handled=errorHandling(dataframe=dataframe)
    dataCount(dataframe=dataframe,how="YES")
    dataCount(dataframe=bd_error_handled,how="NO")
    if os.path.exists("bd_errors.xlsx"):
        with pd.ExcelWriter("bd_errors.xlsx", engine='openpyxl', mode='a') as writer:
            bd_error.to_excel(writer, sheet_name=f"pomiar {file_number+1}")
    else:
        bd_error.to_excel("bd_errors.xlsx", sheet_name=f"pomiar {file_number+1}")
```
