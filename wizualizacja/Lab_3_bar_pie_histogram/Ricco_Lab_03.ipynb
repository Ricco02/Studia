## Wykres słupkowy
Funkcją służącą do tworzenia dowolnych wykresów słupkowych jest `plt.bar(x, y)`, gdzie:
- `x` - lista, macierz lub $n$-krotka współrzędnych $x$ lewych końców podstaw słupków.
- `y` - lista, macierz lub $n$-krotka wysokości słupków.


Wybrane dodatkowe opcje funkcji `bar`:
 - `width` - liczba lub macierz określająca szerokość poszczególnych słupków. Domyślnie 0.8.
 - `bottom` - liczba lub macierz określająca pozycję podstawy słupków na osi $OY$.
 - `color` - kolor słupków (pojedynczy lub w formie listy)
 - `edgecolor` - kolor brzegów prostokątów (pojedynczy lub w formie listy)
 - `linewidth` - grubość brzegów prostokątów. Brak brzegów to 0.
 - `tick_label` - pojedynczy *string* lub ich macierz. Opisy (labele) do znaczników na osiach. Można również użyć osobnych funkcji `plt.xticks(), plt.yticks()`.
 - `xerr` - liczba lub macierz. Błąd względem danych na osi $OX$.
 - `yerr` - liczba lub macierz. Błąd względem danych na osi $OY$.
 - `ecolor}` - kolor (kolory) zakresów błędu.
 - `capsize` - liczba. Długość ,,końcówek'' zakresów błędu.
 - `align` - `'center'` lub `'left'`. Określa wyrównanie słupków.
 - `orientation` - `'vertical'` lub `'horizontal'`. Określa orientację wykresu (pionowy lub poziomy). Wykres słupkowy poziomy można altermatywnie uzyskać za pomocą funkcji `plt.barh()` o analogicznej składni.
 - `log=True/False` - wartość `True` wprowadza skalę logarytmiczną na osiach.

Przykład:

```python
import numpy as np
import matplotlib.pyplot as plt

mSrednia = (28,35,30,35,27)
kSrednia = (37,32,34,20,25)
mOS = (2,3,4,6,2)
kOS = (3,5,2,3,7)
x = np.arange(5)   

p1 = plt.bar(x, mSrednia, width=0.45, color='r',yerr=mOS,capsize=2)
p2 = plt.bar(x, kSrednia, width=0.45,bottom=mSrednia,yerr=kOS,capsize=2)

plt.ylabel('Punkty')
plt.title('Srednia liczba punktow w grupach')
plt.xticks(x, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Mezczyzni', 'Kobiety'))
```

## Histogram

Histogram jest wykresem słupkowym przedstawiającym rozkład danej cechy. Na osi poziomej oznaczone są tzw. przedziały klasowe (wartości) cechy, natomiast na osi pionowej liczba elementów w danym przedziale klasowym.

Do automatycznego obliczenia i utworzenia histogramu z zadanych danych służy funkcja `plt.hist()`. Oprócz wykresu funkcja zwraca w postaci 3-krotki obliczone liczby elementów w przedziale klasowym, kresy przedziałów klasowych i *patche* (w dużym uproszczeniu: rodzaj klasy opisującej wygląd grafiki). Pełna składnia domyślna wygląda następująco:

`plt.hist(x, bins=None, range=None, normed=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, hold=None, data=None)`

gdzie:
 - `x` - dane wejściowe w postaci jednowymiarowej tablicy lub sekwencji takich tablic (niekonieczne tej samej długości `len`). Jedyny nie opcjonalny argument.
 - `bins` - parametr zadający liczbę przedziałów klasowych
 - `range` - zadaje minimalną i maksymalną wartość przedziału klasowego. Domyślne ustawienia to od `x.min()` do `x.max`.
 - `normed=True/False` - jeżeli `True`, pole histogramu zostanie znormalizowane do 1
 - `weights` - `None` lub tablica wag o tych samych wymiarach, co dane `x`
 - `cumulative=True/False` - każdy następny słupek jest sumą poprzednich plus liczba elementów w aktualnym przedziale klasowym
 - `bottom` - tablica o długości równej liczbie przedziałów klasowych, liczba lub `None`. Określa położenie podstawy słupka na wykresie.
 - `histtype` - wygląd histogramu:
    - `‘bar’` - tradycyjny histogram słupkowy. W przypadku kilku zbiorów danych słupki liczby elementów w danym przedziału klasowym są osobne dla każdego zbioru danych.
    - `‘barstacked’` - dla kilku zbiorów danych generuje histogram, w którym liczby wystąpień w danym przedziale klasowym są sumowane dla wszystkich danych i przedstawiane w postaci wspólnego słupka.
    - `‘step’` - niewypełniony obrys histogramu w postaci wykresu liniowego.
    - `‘stepfilled’` - generuje obrys histogramu (wykres liniowy) wraz z wypełnieniem między wykresem a osią poziomą.
 - `align` - kontroluje położenie słupków histogramu względem przedziału klasowego
    - `‘left’` - słupki są wycentrowane na lewym kresie przedziału.
    - `‘mid’` - słupki są wycentrowane w środku przedziału.
    - `‘right’` - słupki są wycentrowane na prawym kresie przedziału.
 - `orientation` - orientacja słupków. Do wyboru pozioma `‘horizontal’` i pionowa `‘vertical’` (domyślnie).
 - `rwidth` - liczba lub `None`. Szerokość słupka relatywnie do długości przedziału klasowego.
 - `log=True/False` - jeżeli `True` skala na osiach będzie skalą logarytmiczną. Puste przedziały klasowe zostaną pominięte.
 - `color` - kolor wykresu. Jeden dla każdego zestawu danych.
 - `label` - pojedynczy *string* lub sekwencja stringów (w przypadku kilku zbiorów danych). Dodaje legendę. Można użyć również funkcji `plt.legend()`.
 - `stacked=True/False` - jeżeli `True`, to wymusza histogram, w którym liczby wystąpień w danym przedziale klasowym są sumowane dla wszystkich danych i przedstawiane w postaci wspólnego słupka, bez względu na wybór typu histogramu `histtype`.

Przykład:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12345)
A=np.random.randn(10000)
np.random.seed(43210)
B=np.random.randn(10000)

plt.subplot(221)
plt.hist(A,bins=30,histtype='step',color='r')

plt.subplot(222)
plt.hist(A,bins=30,histtype='step',orientation='horizontal',color='r')

plt.subplot(223)
plt.hist(A,bins=30,bottom=np.linspace(0,500,30),color='r')

plt.subplot(224)
plt.hist((A,B),stacked=True,rwidth=0.6,color=['r','g'])
plt.legend(['Gauss1','Gauss2'],fontsize='x-small')
plt.tight_layout()
```

## Wykres kołowy

Do tworzenia wykresów kołowych służy funkcja `plt.pie()` o domyślnej składni

`plt.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, hold=None, data=None)`

zwracająca, oprócz wykresu, trójkę list: lista *patchy* opisujących wygląd poszczególnych klinów, lista labeli, lista labeli numerycznych (wartości procentowych w formacie tekstowym).

Opcje funkcji `plt.pie()`:
- `x` - macierz (lista, krotka) danych wejściowych (jedyny wymagany argument funkcji). Kąt rozwarcia klina dla elementu $x_j$ wyniesie $2\pi x_j$`sum(x)` o ile `sum(x)`$>1$. Jeżeli `sum(x)` nie przekracza 1, funkcja `plt.pie()` zinterpretuje każdy element $x_j$ jako od razu zadaną część kąta pełnego. Domyślnie kliny dla poszczególnych danych zaczynają się od godziny 3.00 i układają przeciwnie do ruchu wskazówek zegara.

- `explode` - macierz (lista, krotka) o długości `len(x)` lub `None`. Zadaje wysunięcie względem środka okręgu dla poszczególnych klinów.

- `labels` - lista stringów. Labele dla poszczególnych klinów.

- `colors` - lista kolorów dla poszczególnych klinów.

- `autopct` - pojedynczy *string* zadający format, funkcja lub `None`. Dodaje wewnątrz klinów ich wartość procentową w zadanym formacie lub wywołuje funkcję dodającą właściwy label. Najczęstsze formaty:
     - `'%1.f%%'` - 10% (zaokrąglenie do pełnych procentów, ze znakiem %)
     - `'%1.1f%%'` - 10.0% (jedno miejsce po przecinku, ze znakiem %)
     - `'%1.2f%%'` - 10.00% (dwa miejsca po przecinku, ze znakiem %)
     - `'%1.f'` -10 (zaokrąglenie do pełnych procentów)
     - `'%1.1f'` - 10.0 (jedno miejsce po przecinku)
     - `'%1.2f'` - 10.00 (dwa miejsca po przecinku)
- `pctdistance` - *float* określający położenie tekstu generowanego przez `autopct` względem środka klina.
- `shadow=True/False` - cień (efekt trójwymiarowości)
- `labeldistance` - *float* określający odległość labeli klinów od wykresu.
- `startangle` - *float* określający kąt (w stopniach), na którym zacznie się pierwszy klin.
- `radius` - *float*, promień wykresu.
- `counterclock=True/False` - jeżeli `True`, kliny będą układane przeciwnie do kierunku wskazówek zegara.
- `wedgeprops` - słownik określający własności brzegów klinów i całego wykresu.
- `textprops` - słownik określający własności labeli i innego tekstu.
- `center` - $(x,y)$ określający położenie środka wykresu.
- `frame=True/False` - jeżeli `True`, wyświetli osie wykresu.

Przykład:

```python
import numpy as np
import matplotlib.pyplot as plt

x=(5,10,20,30)
y=(0.3,0,0,0)

plt.pie(x,explode=y,labels=('Ja','Slimaki','Sarny','Dziki'),autopct='%1.2f%%',shadow=True,startangle=30)
plt.title('Kto zjadl moje truskawki?')
```
